```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Fetch TSLA stock prices using Yahoo Finance
def fetch_stock_prices(symbol, start_date, end_date):
    data = yf.download(symbol, start=start_date, end=end_date)
    return data


# Calculate price direction
def calculate_price_direction(prices):
    return np.where(prices.diff() > 0, 1, 0)

# Create features and labels
def create_features_and_labels(data, window_size=7):
    features = []
    labels = []
    for i in range(len(data) - window_size):
        window = data[i:i+window_size]
        features.append(window.values)
        labels.append(data.iloc[i+window_size])
    return np.array(features), np.array(labels)

# Fetch TSLA stock prices
symbol = 'TEL'
start_date = '1990-01-01'
end_date = '2024-01-01'
data = fetch_stock_prices(symbol, start_date, end_date)
display(data)
# Calculate price direction
data['price_direction'] = calculate_price_direction(data['Close'])
display(data)
data.to_csv('direction.csv')
chunks = [data['price_direction'][i:i+15] for i in range(0, len(data), 15)]
```
